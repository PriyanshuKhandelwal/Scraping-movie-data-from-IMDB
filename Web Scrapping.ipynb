## Using request.get to receive response from server using the imdb url:

```python
import requests
url = 'https://www.imdb.com/search/title?release_date=2018&sort=num_votes,desc&page=2&ref_=adv_nxt'
headers = {"Accept-Language": "en-US, en;q=0.5"}
response = requests.get(url)
print(response.text)
```

` Above is the response we got from the url we used  in requests.get()`

Let's now convert the result to `BeautifulSoup` format using **BeautifulSoup** and **html.parser**

```python
from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, "html.parser")
type(html_soup)
```

# Full code begins from here:

```python
from requests import get
import warnings
import time
from IPython.core.display import clear_output
from bs4 import BeautifulSoup
from time import sleep
from random import randint

names=[]
years=[]
imdb_ratings=[]
meta_ratings=[]
votes=[]

start_time = time.time()
requests = 0
headers = {"Accept-Language": "en-US, en;q=0.5"}
title_count = 1
while title_count<=400:
    url = "https://www.imdb.com/search/title?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&start={}&ref_=adv_nxt".format(title_count)
    response = get(url, headers=headers)
    sleep(randint(8,15))
    requests += 1
    elapsed_time=time.time() - start_time
    print("Request:{} Frequency: {} requests/second".format(requests, requests/elapsed_time))
    clear_output(wait=True)
    
    if response.status_code != 200:
        warnings.warn("Request: {}, Status Code:{}".format(requests,response.status_code))
    
    if requests>8:
        warnings.warn("Number of requests are greater than expected.")
        break
        
    html_soup = BeautifulSoup(response.text,'html.parser')
    
    my_containers=html_soup.find_all("div", class_="lister-item mode-advanced")
    
    for container in my_containers:
        if container.find("div", class_="ratings-metascore") is not None:
            name = container.h3.a.text
            names.append(name)
        
            year = container.h3.find("span", class_="lister-item-year text-muted unbold")
            year = year.text
            years.append(year)
            
            imdb = float(container.strong.text)
            imdb_ratings.append(imdb)
            
            meta = container.find("span", class_ = "metascore").text
            meta_ratings.append(meta)
            
            vote = int(container.find("span",attrs={"name":"nv"})["data-value"])
            votes.append(vote)
    
    
    title_count+=50
   
```

```python
import pandas as pd
movie_ratings = pd.DataFrame({'movie':names,
                              'year':years,
                              'imdb_rating':imdb_ratings,
                              'meta_rating':meta_ratings,
                              'votes':votes})
```
